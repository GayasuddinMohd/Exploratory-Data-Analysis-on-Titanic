```python
# Create a function to detect if a person is a man, woman or child.
def man_wom_chi(passenger):
    age=passenger['Age']
    sex=passenger['Sex']
    
    return 'child' if age < 16 else sex
```

```python
# Create a new column "Person" in the dataset.
titanic_df['Person']=titanic_df.apply(man_wom_chi,axis=1)
```

```python
# Looking at the first five records of the datatset.
titanic_df.head()
```

# Which deck has most number of passengers ?

```python
# Create a dataset deck_df and we just drop the missing data in Cabin column.
deck_df=titanic_df.dropna(axis=0)
deck_df
```

```python
# Head of the datatset.
deck_df.head()
```

```python
# Create a function to extract level from the Cabin column.
def get_level(passenger):
    cabin=passenger['Cabin']
    return cabin[0]
```

```python
# To avoid settingWithCopyWarning.
pd.options.mode.chained_assignment=None
```

```python
# Create a new column level in the deck_df column.
deck_df['level']=deck_df.apply(get_level,axis=1)
```

```python
# Head of the datatset after adding new column level in it.
deck_df.head()
```

```python
# Catplot for finding which deck has most number of passengers.
sns.catplot(x='level',data=deck_df,kind='count',ec='black',lw=2)
plt.title("DECK",fontsize=18)
plt.xlabel("Name of the Deck",fontsize=13)
plt.ylabel("Number of Passengers",fontsize=13)
plt.grid(True)
plt.show()
```

From the above, we can say that most of the passengers belong to deck C on Titanic.

# Relation between the class and Deck of the passengers ?

```python
# Catplot bw class and deck of the passengers.
sns.catplot(x='level',data=deck_df,kind='count',hue='Pclass',ec='black',lw=2)
plt.title("DECK VS CLASS",fontsize=18)
plt.xlabel("Name of the Deck",fontsize=13)
plt.ylabel("Number of Passengers",fontsize=13)
plt.grid(True)
plt.show()
```

From the above, we can say that the upper Deck(A,B,C) has only passengers from the class 1.
And the passengers from class 3rd start from lower Deck E.

# How many passengers were with their family members on the Titanic ?

```python
# Count of the Sibsp column of the datatset.Sibsp column contains data of number of siblings or spouse of the passenger. 
titanic_df.SibSp 
```

```python
# Count of the Parch column of the datatset.Parch column contains data of number of parents or childrens of the passenger.
titanic_df.Parch
```

```python
# Tail of the datatset.
titanic_df.tail()
```

```python
# Where Alone has 0 value assign Yes and where Alone has value greater than 1 assign No.
titanic_df['Alone'].loc[titanic_df['Alone']>0] = 'No'
titanic_df['Alone'].loc[titanic_df['Alone']==0] = 'Yes'
```

```python
# Head of the datatset.
titanic_df.head()
```

```python
# Catplot whether the passenger is travelling with family or without family members.
sns.catplot(data=titanic_df,x='Alone',kind='count',ec='black',lw=2)
plt.title("FAMILY",fontsize=18)
plt.grid(True)
plt.show()
```

From the above we can say that, majority of the passengers on the ship were with their family members.

# Which class of the passengers survived most on the Titanic ?

```python
# Cat-plot about class of the passengers who survived.
sns.catplot(data = titanic_df,x ='Pclass',y='Survived',kind='point')
plt.grid(True)
plt.ylabel("Survived",fontsize=13)
plt.xlabel("Class of the Passengers",fontsize=13)
plt.show()
```

From the above we can say that most of the passengers are survived from class 1 because they have higher Deck on the ship, and there are very few passengers from class 3 who survived because they have lower Deck on the ship.

# Which one male,female or child of what class Survived has less chance to survived on the Titanic ?

```python
# Cat-plot about class and gender of the passengers who have less chances of survival.
sns.catplot(data=titanic_df,x='Pclass',y='Survived',kind='point',hue='Person')
plt.grid(True)
plt.ylabel("Survived",fontsize=13)
plt.xlabel("Class of the Passengers",fontsize=13)
plt.show()
```

# Is there any realtion between the age and survival of the passengers on the Titanic ?

```python
# Create lmplot to check the chances of survival of the passengers on the basis of their age.
sns.lmplot(data=titanic_df,x='Age',y='Survived')
plt.ylabel("Survived",fontsize=13)
plt.xlabel("Age of the Passengers",fontsize=13)
plt.grid(True)
plt.show()
```

From the above we can say that, the older passengers has very less chances of survival as compared to the younger passengers.

# Check the chances of survival of the passengers on the basis of their age and class ?

```python
# Create lmplot to check the chances of survival of the passengers on the basis of their age and class.
sns.lmplot(data=titanic_df,x='Age',y='Survived',hue='Pclass')
plt.ylabel("Survived",fontsize=13)
plt.xlabel("Age of the Passengers",fontsize=13)
plt.grid(True)
plt.show()
```

From the above we can say that, the passengers who have younger in age and of class 1 has greater chances of survival than that of who have older age and of class 3.
